```python
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
# fit model

model = ARIMA(df, order=(7,1,0))
model_fit = model.fit(disp=0)
print(model_fit.summary())
```

```python
output = model_fit.forecast()
```

```python
output[0]
```

```python
# plot residual errors
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
plt.show()
residuals.plot(kind='kde')
plt.show()
print(residuals.describe())
```

### Start to use ARIMA model to predict

## ARIMA

```python
#Each time a new value Y is predicted, add the new y into x as an actual value to predict the next Y.
X = df.values
size = int(len(X) * 0.9)
train, test = X[:size], X[size:]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
    model = ARIMA(history, order=(7,1,0))
    model_fit = model.fit(disp=False,  trend='nc')
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
```

```python
predictions
```

```python
len(predictions)
```

```python
error = np.sqrt(mean_squared_error(test, predictions))
print('Test RMSE: %.3f' % error)
```

```python
# plot
plt.plot(test)
plt.plot(predictions, color='red')
plt.show()
```

When predicting future sales, using the ARIMA model would place too much weight on the sales of the previous day, which may lead to inaccuracies.
